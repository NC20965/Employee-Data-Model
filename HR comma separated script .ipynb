# Explore the variables of the hr data set that can help in identifying the factors that led some employees to leave the company



## Average montly hours and data type conversion


```python
#data typ conversion
df["average_montly_hours"] =pd.to_numeric(df["average_montly_hours"])

df.head()
```

## Get maximum hours per department

```python
Get_max = df["average_montly_hours"].max()
print(Get_max)
```

## Get minimun hours per department

```python
Get_min = df["average_montly_hours"].min()
print(Get_min)
```

## Get department name with MAXIMUM monthly hours

```python
#use Arrow function to get department name:
df[">= 310"] = df["average_montly_hours"].apply(lambda Department: "True" if Department >=310  else "False")
 
print(df[">= 310"])

# Define a function to get department name
def get_department(average_montly_hours,Department):
    average_montly_hours = pd.to_numeric(average_montly_hours)
    
    count =0
    Found_Department = []
    if average_montly_hours==310:
        count =0
        for j in df.iterrows():
            Department = Found_Department.append(df.Department)
            if count ==0:
                print("{} works more monthly hours".format(Department))
            else:
                pass
              
    else:
        pass
    
get_department(310,"Department")
```

## Department with MAX average monthly hours

```python
df[(df.average_montly_hours==310) & (df.Department=="hr")]
#print()
```

## Count number of low payments in in salary column

```python
df = pd.read_csv("HR_Comma_Sep_Assessment.csv")
Result=df[(df.salary=="low") & (df.Department =="hr")].count()
print(Result)
df.head()
```

```python
#Bar graph plot
satisfaction_level=range(0 ,1)
plt.scatter(satisfaction_level, Result["salary"])
#plt.xticks(df["salary"])
plt.ylabel("Satisfaction levels")
plt.xlabel("Salary range")
plt.colorbar()
plt.title('The level of HR empoye sastifaction')
plt.show()
```
